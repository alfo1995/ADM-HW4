### Degree centrality



**which is defined as the number of links incident upon a node**

```python
modules.loglog_plot(H)
```

In the log-log plots shown above we can see how Riemannâs Z function, generalization of the Zipf function, and the Pareto distribution (the continuous equivalent of the Zipf distribution) are those that best approximate our data.

Since our network is not completely random, the Poisson distribution is not a good fit for the data.

### Betweenness centrality

**Betweenness centrality quantifies the number of times a node acts as a bridge along the shortest path between two other nodes.**

**Given that the output *nx.betweenness_centrality(G)* is a dictionary with keys equals to the nodes of the subgraph and values equals to the corresponding betweenness centrality, we create a function called *most_important* which returns the nodes with highest betweenness centrality measure.
**

** So higher the betweenness of a node more control this node has over the network**

```python
Gt= modules.most_important(H)
modules.plot_betwenness(Gt)
```

### Closeness centrality



**In a connected graph, the normalized closeness centrality (or closeness) of a node is the average length of the shortest path between the node and all other nodes in the graph. Thus the more central a node is, the closer it is to all other nodes.**




*Even for the closeness centrality we create a function which returns the top ten nodes with highest measure, and we visualize the subgraph induced by this node from the starting subgraph*

```python
modules.top_clos_node(H)
```
